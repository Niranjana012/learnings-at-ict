## 10.Conditional Aggregates
 Calculate the mean mpg for cars grouped by the number of cylinders using NumPy 
 techniques.

```python
cyl1=df1["cylinders"].unique()
```

```python
for c in cyl1:
    part=df1[df1["cylinders"]==c]
    avg=part["mpg"].mean()
    print("average mean mpg of {} cylinders grouped is:{}".format(c,avg))
```
