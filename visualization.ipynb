## Bar Graph

```python
x = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
y = [22,55,66,33,88,44,99]
plt.bar(x, y)
plt.xlabel("Days")
plt.ylabel("Counts")
plt.show()
```

```python
x = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
item1sales = [22,55,66,33,88,44,99]
item2sales = [43,56,75,32,84,21,12]
plt.bar(x,item1sales, color="y", label="items1")
plt.bar(x,item2sales, color="b",alpha=.2, label="item2")
plt.xlabel("Days")
plt.ylabel("Counts")
plt.legend()
plt.show()
#alpha=0(invisible fully transparent)
#alpha=1(opaque)
#alpha=.2(low transparency)
```

## BOX PLOT

```python
a = np.random.normal(170, 1, 20000)
q1 = np.quantile(a, 0.25)
q3 = np.quantile(a, 0.75)
iqr = q3-q1
lb = q1-(1.5*iqr)
ub = q3+(1.5*iqr)
sns.boxplot(a)
```

```python
from scipy import stats
import matplotlib.pyplot as plt
data = np.random.exponential(scale=1,size=1000)
tranformed_data, lambda_valu = stats.boxcox(data)

plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.hist(data, bins=30)
plt.title('Orginal Data')

plt.subplot(1,2,2)
plt.hist(tranformed_data, bins=30)
plt.title('Box-Cox Transformed Data')

plt.show()
```
